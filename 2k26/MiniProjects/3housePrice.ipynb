### ***HOUSE PRICE PREDICTION***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
housedata = sklearn.datasets.fetch_california_housing()
print(housedata)
```

```python
df = pd.DataFrame(housedata.data, columns=housedata.feature_names)
```

```python
df
```

```python
#add the target(price) column to the DataFrame
df['Price'] = housedata.target
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#statistical measures of the dataset
df.describe()
```

***Understanding the correlation between various features in the dataset:***
* ***Posetive Correlation***
* ***Negative Correlation***

```python
#construction a HEATMAP to understand the correlation:
correlation = df.corr()
plt.Figure(figsize=(10,10))
sns.heatmap(correlation, 
            cbar=True, 
            square=True, 
            fmt='.1f',  #decimal point
            annot=True, #numerals
            cbar_kws={'shrink': 0.8},
            cmap='Blues')
plt.show()
```

***Splitting the Data and Target:***

```python
X = df.drop(['Price'], axis=1)
X
```

```python
y = df['Price']
y
```

***Splitting the Data into Training Data and Test Data:***

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

***Model Training:***

```python
#loading the model
xgbmodel = XGBRegressor()
```

```python
#training the model with x_train
xgbmodel.fit(X_train, y_train)
```

***Evaluation:***

* *Prediction on training data*

```python
#accuracy for prediction on training data
training_data_prediction = xgbmodel.predict(X_train)
```

```python
training_data_prediction
```

```python
#R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)
print("R Squared Error : ", score_1)
```

```python
#Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)
print("Mean Absolute Error : ", score_2)
```

***Train Data :*** *Visualizing the actual Prices and Predicted prices*

```python
sns.scatterplot(x=y_train, y=training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Prediction Price")
plt.title("Actual Price vs Predicted Price")
```

***Prediction on Test Data:***

```python
#accuracy for prediction on test data
testing_data_prediction = xgbmodel.predict(X_test)
testing_data_prediction
```

```python
#R squared error
score_1 = metrics.r2_score(y_test, testing_data_prediction)
print("R Squared Error : ", score_1)
```

```python
#Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_test, testing_data_prediction)
print("Mean Absolute Error : ", score_2)
```

***Test Data :*** *Visualizing the actual Prices and Predicted prices*

```python
sns.scatterplot(x=y_test, y=testing_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Prediction Price")
plt.title("Actual Price vs Predicted Price")
```
