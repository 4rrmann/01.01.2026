***q10: how many country have Republic in their full name***

```python
df['country_long']
```

```python
count = 0
def counting(txt):
    global count              #use global counter to access within func
    if 'republic' in txt.lower():  #case-insensitive check {lower-case alphabet}
        count += 1            
    return txt                

df['country_long'] = df['country_long'].apply(counting)
count
```
