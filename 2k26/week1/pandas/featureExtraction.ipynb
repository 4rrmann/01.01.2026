***q1: make a new column for episode count***

```python
df.loc[1]['Title']
```

```python
# (..)
def extract_episodes(txt):
    check = False   #inside parentheses?
    data = ""       #collected text

    for i in txt:   
        if i == ")":        #stop at ')'
            check = False
            return data
        if check:           #if inside '( )'
            data += i
        if i == '(':        #start after '('
            check = True
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
df['Episodes'] = df['Episodes'].str.replace('eps', '')
```

***q2: make a new column for time stamp***

```python
df.loc[1]['Title']
```

```python
def extraction_time(txt):
    # check = False   #flag (unused for now)
    data = ""       #stores extracted text

    for i in range(len(txt)):          #scan entire string
        if txt[i] == ')':              #start after ')'
            for j in range(i+1, i+20): #grab next 20 chars
                data += txt[j]         #append to data
            return data                #return result
```

```python
df['Time Stamp'] = df['Title'].apply(extraction_time)
```

```python
df.head()
```

```python
#
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')     #split date range
        start_date = datetime.strptime(start_str, '%b %Y')  #parse start
        end_date = datetime.strptime(end_str, '%b %Y')      #parse end
        r = relativedelta(end_date, start_date)       #date difference
        return r.years * 12 + r.months + 1            #total months (+start)
    except:
        return None    #invalid format

df['Months'] = df['Time Stamp'].apply(calculate_total_months)  #apply to column
```
