```python
import numpy as np
import pandas as pd
```

***Merging 2 dataframes***

```python
# DataFrame 1: Employee information
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

employees
```

```python
# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})

salaries
```

```python
pd.merge(employees,salaries, on='employee_id', how='inner')
```

```python
pd.merge(employees,salaries, on='employee_id', how='outer')
```

```python
#on the basis of left{employees}
pd.merge(employees,salaries, on='employee_id', how='left')
```

```python
#on the basis of right{salaries}
pd.merge(employees,salaries, on='employee_id', how='right')
```

***Concatination of 2 dataframes***

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df1
```

```python
df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})

df2
```

```python
#on the basis of columns{by default asix=0}:
pd.concat([df1,df2])
```

```python
#on the basis of rows:
pd.concat([df1,df2], axis=1)
```

***Joining 2 data frames***

```python
# First DataFrame:
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

df1
```

```python
# Second DataFrame:
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])

df2
```

```python
df1.join(df2) #joining on the basis of df1
```

```python
df2.join(df1) #joining on the basis of df2
```

```python
df1.join(df2, how='outer') #joining {including everything}
```
