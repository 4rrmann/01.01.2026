```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

***Heatmap:***
<br>*a heatmap is used to **visualize numerical values in a matrix form using color intensity, commonly for correlations**, patterns, and comparisons across two dimensions.*

```python
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
corr = tipscorr.corr()
corr
```

```python
sns.heatmap(
    corr,                #data
    annot=True,          # display correlation values inside each cell
    cmap="coolwarm",     # distinguish positive and negative correlations
    center=0,            # keep 0 correlation as the neutral color
    linewidths=0.3       # add lines between cells for better clarity
)
```

***Cluster map:***
<br>*a clustermap **visualizes a heatmap with hierarchical clustering**, grouping similar rows and columns to reveal hidden patterns and relationships in the data.*

```python
sns.clustermap(
    corr,                    #data
    annot=True,              # show correlation values
    cmap="coolwarm",         # clear positive/negative contrast
    center=0,                # center color at 0 correlation
    linewidths=0.8,          # separate cells clearly
    figsize=(6, 5),          # better readability
    fmt=".2f"                # round values
)
```

***Pivot table Heatmap:***

```python
pvtFlight = flights.pivot_table(values='passengers',
                                index='month',
                                columns='year', observed=False)

pvtFlight
```

```python
sns.heatmap(pvtFlight)
```
