```python
import matplotlib.pyplot as plt
import seaborn as sns
```

*a regression plot `(regplot/lmplot)` visualizes the relationship between two numerical variables by displaying a scatter of points along with **a best-fit regression line** and its confidence interval.*
* *regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it*

```python
tips = sns.load_dataset('tips')
tips
```

***y: predictive target*
<br>*x: through which prediction happens***

***algo: linear regg***

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s' :30})
```
