```python
import matplotlib.pyplot as plt
import seaborn as sns
```

***Categorical Data Plots:***
<br>*Categorical plots are graphs used to **visualize and compare data across discrete categories** (like groups or labels), typically used when the **x-axis represents categories** rather than continuous values.*

*Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:*
* ***boxplot***
* ***violinplot***
* ***stripplot***
* ***swarmplot***
* ***barplot***
* ***countplot***

```python
df = sns.load_dataset('tips')
df
```

***Countplot:***
<br>*a countplot is a categorical plot that displays the **number of observations**/**frequency (count) in each category.***

```python
sns.countplot(x=df['sex'], hue=df['smoker'])
```

***Barplot:***
<br>*a barplot is a categorical plot that shows the **average (or another estimator) of a numerical variable for each category**, with optional confidence intervals.*

```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

*you can **change the estimator object to your own function**, that converts a vector to a scalar,
<br>**Mean (default):** np.mean
<br>**Median:** np.median
<br>**Sum:** np.sum
<br>**Count:** len (to count the number of observations)
<br>**Standard Deviation:** np.std Minimum: np.min Maximum: np.max*

***

***Boxplot:***
<br>*a boxplot is a categorical plot that visualizes the **distribution of a numerical variable across categories using median, quartiles, and outliers.***

```python
sns.boxplot(x='tip', y='day',hue='day', data=df, palette='rainbow')
```

***boxplot and violinplot:***
<br>*boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.*

***Violinplot:***

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

***

***Stripplot and Swarmplot:***
<br>The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x='day', y='total_bill', data=df)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
#combo:
sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df, size=4)
```
