```python
import seaborn as sns
import matplotlib.pyplot as plt
```

*We use **distribution plots** to visualize the **distribution of quantitative data.***

*Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:*

* ***distplot/histplot***
* ***jointplot***
* ***pairplot***
* ***rugplot***

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

*distplot/**Histplot:***
<br>*checks the distribution of the data.*

```python
plt.subplot(1,2, 1)
sns.histplot(df['total_bill'], bins=20, kde=True)
#kde: kernal density estimeter
plt.subplot(1,2, 2)
sns.histplot(df['tip'], bins=20, kde=True)
```

***Jointplot:***
<br>*checks the distribution bw two quantitative data points.*

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

****Pairplot:***
<br>*A pair plot visualizes **pairwise relationships between multiple numerical variables**, showing **scatter plots for variable pairs** and **distributions on the diagonal** for quick exploratory data analysis.*

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='size', palette='rainbow')
```

***Rugplot:***

```python
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(df)
```
