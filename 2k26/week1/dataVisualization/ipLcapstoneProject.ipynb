```python
best = (
    df.assign(
        wickets=df['best_bowling_figure'].str.split('--').str[0].astype(int),
        runsbywickets=df['best_bowling_figure'].str.split('--').str[1].astype(int)
    )
    .sort_values(['wickets', 'runsbywickets'], ascending=[False, True])
    .iloc[0]
)

best[['best_bowling', 'best_bowling_figure']]
```

```python
best = (
    df.assign(
        # Create a new column 'wickets'
        # Split '5--10' â†’ ['5', '10'] â†’ take index 0 â†’ convert to int
        wickets=df['best_bowling_figure'].str.split('--').str[0].astype(int),

        # Create a new column 'bbf' (runs conceded)
        # Split '5--10' â†’ ['5', '10'] â†’ take index 1 â†’ convert to int
        runsbywickets=df['best_bowling_figure'].str.split('--').str[1].astype(int)
    )
    # Sort primarily by wickets (descending â†’ highest first)
    # If wickets are same, sort by runs conceded (ascending â†’ lowest first)
    .sort_values(['wickets', 'runsbywickets'], ascending=[False, True])

    # Pick the first row after sorting (best bowling figure)
    .iloc[0]
)

# Display only required columns
best[['best_bowling', 'best_bowling_figure']]
```

```python
df.loc[
    df['best_bowling_figure']
      .str.split('--')
      .apply(lambda x: (int(x[0]), -int(x[1])))
      .idxmax(),
    ['best_bowling', 'best_bowling_figure']
]
```

```python
df.loc[
    # Split 'best_bowling_figure' like '5--10' into ['5', '10']
    df['best_bowling_figure']
      .str.split('--')

      # Convert each split value into a tuple:
      # (wickets, -runs_conceded)
      # - wickets â†’ higher is better
      # - runs â†’ negated so that fewer runs become a larger value
      .apply(lambda x: (int(x[0]), -int(x[1])))

      # Find the index of the "maximum" tuple
      # Python compares tuples element-wise:
      # first wickets (max), then runs (min due to negative)
      .idxmax(),

    # Select only the required columns from that row
    ['best_bowling', 'best_bowling_figure']
]
```
