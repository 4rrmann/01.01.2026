***basic commands***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5, 11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("x / y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

***subplots:***

```python
plt.subplot(2,2,1)
plt.plot(x**2,x**2)

plt.subplot(2,2,2)
plt.plot(y,x)

plt.subplot(2,2,3)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(x,y**3)
```

***object oriented way***

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1, 1,1])
axis1.plot(x,y)

axis2 = fig.add_axes([0.5,0.3, 0.4,0.5])
axis2.plot(x,y)
```

***types of plots:***

```python
#scatter plot:
plt.scatter(x,y)
```

```python
#histogram:
from random import sample
data = sample(range(1,1000), 100)
plt.hist(data)
```

```python
#box plot:
plt.boxplot(x)
```

***saving Plots:***

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1, 0.8,0.8])
ax.plot(x,y)
plt.savefig("saved.png")
```

```python
xaxis = np.linspace(0,6,11) #np.linspace(start,stop, nums bw them{evenly spaced})
```

```python
fig, ax = plt.subplots(figsize=(12,6)) #width:12inces, height:6inces
```

```python
fig, ax = plt.subplots(figsize=(12,6)) #width:12inces, height:6inces

#plots a line(x, y= xaxis+n {shifts each line upwards}, line color, linewidth controls thickness of the line)
ax.plot(xaxis, xaxis+1, color="red", linewidth=0.25)
ax.plot(xaxis, xaxis+2, color="red", linewidth=0.50)
ax.plot(xaxis, xaxis+3, color="red", linewidth=1.00)
ax.plot(xaxis, xaxis+4, color="red", linewidth=2.00)
```

```python
fig, ax = plt.subplots(figsize=(12,6)) #width:12inces, height:6inces

# possible linestype options '-', '-.', -., ':', 'steps'

#solid line: '-'
ax.plot(xaxis, xaxis+5, color="green", lw=3, linestyle='-')

#dash-dot line: '-.'
ax.plot(xaxis, xaxis+6, color="green", lw=3, ls='-.')

#dotted line: ':'
ax.plot(xaxis, xaxis+7, color="green", lw=3, ls=':')
```

```python
fig, ax = plt.subplots(figsize=(12,6)) #width:12inces, height:6inces

#custom dash line:
line, = ax.plot(xaxis, xaxis+8, color="black", lw=1.50)
line.set_dashes([5,50, 15,50]) #format: line length(ll),space length(sl), ll,sl, ...
```

```python
fig, ax = plt.subplots(figsize=(12,6)) #width:12inces, height:6inces

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(xaxis, xaxis+ 9, color="blue", lw=1, ls='-', marker='+')  #plus sign
ax.plot(xaxis, xaxis+10, color="blue", lw=2, ls='--', marker='o') #circle
ax.plot(xaxis, xaxis+11, color="blue", lw=2, ls='-', marker='s')  #square
ax.plot(xaxis, xaxis+12, color="blue", lw=2, ls='--', marker='1') #tri-dot

# ls + marker shows line + point representation
```

```python
fig, ax = plt.subplots(figsize=(12,6)) #width:12inces, height:6inces

ax.plot(xaxis, xaxis+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(xaxis, xaxis+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(xaxis, xaxis+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
```

```python
fig, ax = plt.subplots(figsize=(12,6))  #width:12inces, height:6inces

#a bit adv marker customization:
ax.plot(xaxis, xaxis+16,
        color="purple",
        lw=1, ls='-',
        marker='s', markersize=8, 
        markerfacecolor="yellow",
        markeredgewidth=3, markeredgecolor="green")
```

*all in one:*

```python
fig, ax = plt.subplots(figsize=(12,6))  #width:12inces, height:6inces

#plots a line(x, y= xaxis+n {shifts each line upwards}, line color, linewidth controls thickness of the line)
ax.plot(xaxis, xaxis+1, color="red", linewidth=0.25)
ax.plot(xaxis, xaxis+2, color="red", linewidth=0.50)
ax.plot(xaxis, xaxis+3, color="red", linewidth=1.00)
ax.plot(xaxis, xaxis+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(xaxis, xaxis+5, color="green", lw=3, linestyle='-')
ax.plot(xaxis, xaxis+6, color="green", lw=3, ls='-.')
ax.plot(xaxis, xaxis+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(xaxis, xaxis+8, color="black", lw=1.50)
line.set_dashes([5, 50, 15, 50]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(xaxis, xaxis+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(xaxis, xaxis+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(xaxis, xaxis+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(xaxis, xaxis+12, color="blue", lw=3, ls='--', marker='1')
ax.plot(xaxis, xaxis+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(xaxis, xaxis+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(xaxis, xaxis+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(xaxis, xaxis+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
