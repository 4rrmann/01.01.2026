```python
import numpy as np
```

***Numpy indexing & Slicing of vectors***

```python
arr = np.arange(11,21)
arr
```

```python
arr[6] #index
```

```python
#slicing
arr[1:6]
```

```python
arr[:6] #by default it takes 0{as initial index}
```

```python
arr[:] #by default: 0th initial index - last final index
```

```python
#steps slicing
arr[4::2]
```

***Numpy indexing & Slicing of Matrix***

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[3] #rows indexing {starting from 0}
```

```python
arr[3,4] #i'th row j'th column
```

```python
slice_arr = arr[3:, 3:]
slice_arr
```

```python
arr[:, 4] #column indexing {starting from 0} -> now this a vector
```

***Boolean indexing***

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = (arr%2 == 0)
bool_index
```

```python
bool_arr = arr[bool_index]
bool_arr
```
