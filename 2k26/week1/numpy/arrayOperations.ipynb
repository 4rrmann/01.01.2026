***Deep & Shallow copy***

```python
a = np.arange(1,21)
a
```

```python
slice_arr = a[:5]
slice_arr *=10
slice_arr
```

```python
a #slice{shallow copy of a} -> no changes occur in 'a'
```

```python
b = a
```

```python
b[0] = 77 #now a changes -> deep copy
a
```

```python
#copy
b = a.copy()
b[0] = 7
```

```python
b
```

```python
a
```

***Advance Array Manipulation***

*Stacking Array:*

```python
a = np.arange(1,5)
a
```

```python
b = np.arange(5,9)
b
```

```python
#vertical stack
np.vstack((a,b)) #{length of vectors/arrays should be same}
```

```python
#horizontal stack
np.hstack((a,b)) #{join}
```

```python
#column stack
np.column_stack((a,b))
```

*Splitting Array:*

```python
s = np.arange(16).reshape(4,4)
s
```

```python
np.hsplit(s,2) #s{array}, 2{horizontal splits}
```

```python
np.vsplit(s,2) #s{array}, 2{vertically splits}
```
