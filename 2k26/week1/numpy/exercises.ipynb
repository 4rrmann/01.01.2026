***Valid Sudoku***

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
s
```

```python
np.sum(s,axis=1) #sum of all rows individually
```

```python
np.sum(s, axis=0) #sum of all columns individually
```

```python
#extracted the 1st block of Sudoku
s[:3, :3]
```

```python
#extracted the 2nd block of Sudoku
s[:3, 3:6]
```

```python
#extracted the 3rd block of Sudoku
s[:3, 6:9]
```

Optimized solution:

```python
for i in range(0,9, 3):
    for j in range(0,9, 3):
        b = s[i:i+3, j:j+3]
        # print(b)
        print(b.sum())
```

***General ques***

```python
# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix
data.shape
```

```python
#Find the average age of students
np.mean(data[:,0])
```

```python
#Extract Math marks of all students
data[:, 1]
```

```python
#Find the highest Science mark
data[:, 2].max()
```

```python
#Get details of the student who scored more than 90 in Math
x = data[:, 1] > 90
data[x]
```

```python
#Increase Math marks of all students by 5
data[:, 1] + 5
```

```python
#Find how many students are younger than 19.
y = data[:, 0] < 19
younger = data[y]
print(f"younger students data: \n {younger}")
```

```python
print(f"{len(data[y])} students are younger than 19")
```

```python
#Calculate the average marks in each subject (column-wise mean)
maths = data[:, 1:2]
maths
```

```python
maths.mean()
```

```python
sci = data[:, 2:]
sci
```

```python
sci.mean()
```

```python
#compact version of average marks for maths, sci.
np.mean(data[:,1:], axis=0)
```

```python
#Get data of students who scored at least 80 in both subjects
m = data[:, 1] >= 80
s = data[:, 2] >=80
data[m & s]
```
