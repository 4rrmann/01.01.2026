```python
import numpy as np 
from scipy.stats import norm 
```

*for a z-test, the usual assumption is sample size ≥ 30 (or known population σ)*

```python
#for 
sample = [
    168, 171, 172, 169, 170,
    173, 174, 167, 169, 171,
    172, 170, 168, 169, 173,
    174, 175, 170, 171, 169,
    168, 172, 173, 170, 171,
    169, 168, 174, 175, 170,
    171, 172, 169, 170, 173
]
```

```python
population_mean = 170 #given
population_std = 3    #given

samp_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (samp_mean - population_mean) / (population_std / np.sqrt(n))
z_score
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
p_value
```

```python
alpha = 0.05

if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("i will accept the null hypothesis ")
```
